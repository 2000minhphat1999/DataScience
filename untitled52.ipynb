```python
from phonemizer import phonemize

text = "Xin chào thế giới!"
phonemes = phonemize(text, language='vi', backend='espeak')
print(phonemes)
```

```python
from TTS.api import TTS

# Khởi tạo TTS với mô hình đã tải xuống
tts = TTS(model_name="tts_models/vi/vctk/vits", progress_bar=False)

# Tổng hợp giọng nói từ chuỗi âm vị
tts.tts_to_file(text=phonemes, file_path="output.wav")
```
